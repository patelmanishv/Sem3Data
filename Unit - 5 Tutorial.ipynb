## WAP to find the length of the longest consecutive members in a given list

```
input = [-2,-4,4,2,3,1,0]
output = 5

input = [6,4,2,3,1]
output = 4
```

````python
## Solution by Krish
# l1 = [-2,-4,4,2,3,1,0]
l1 = [0,1,2,3,5,7,6,100,8,9,8]
l1.sort()
count = max = 0
for i in range(len(l1) - 1):
    if l1[i] == l1[i+1] - 1:
        count+=1
        if count>max:
            max = count
    elif l1[i] == l1[i+1]:
        continue
    else:
        count = 0
print(max+1)
````

## alternate solution - with set

````python
def longest(nums):
    result = 0
    nums = set(nums)
    for num in nums:
        if num-1 not in nums:
            current = num
            length = 1
            while current+1 in nums:
                current+=1
                length+=1
            result = max(result,length)
    return result
````

## For a given list of positive integers A, let S be sum of the digits of minimum element of A; Return 0 if S is odd otherwise 1.

```
input = [34,33,1,24,75,33,54,8]
output = 0
Explanation - 1 is minimum and is odd hence 0

input = [99,77,33,66,55]
output = 1
Explanation = 33 is minimum and sum of digits is 6 which is even; hence 1
```

````python
def sumdigits(A):
    mini = min(A)
    result = 0
    while mini > 0:
        quo = mini % 10
        rem = mini // 10
        result += quo
        mini = rem
    return 1 if result % 2 == 0 else 0
````

````python
sumdigits([99,77,33,66,55])
````

## PB 448
```
Given a list L of size N. You need to count the number of special elements in the given list. An element is special if removal 
of that element makes the list balanced. The list will be balanced if sum of even index elements is equal to the sum of odd 
elements. Also print the updated lists after removal of special elements.
 Example 1:
 Input: 
L=[5, 5, 2, 5, 8]
 Output:
 Original List: [5, 5, 2, 5, 8]
 Index to be removed is: 0
 List after removing index 0 : [5, 2, 5, 8]
 Original List: [5, 5, 2, 5, 8]
 Index to be removed is: 1
 List after removing index 1 : [5, 2, 5, 8]
 Total number of special elements: 2
 Explaination:
 If we delete L[0] or L[1], list will be balanced.
 [5, 2, 5, 8]
 (5+5) = (2+8)
 So L[0] and L[1] are special elements, So Count is 2.
 After removal of the special elements, list will be: [5, 2, 5, 8]
 Example 2:
 Input:
 L=[2,1,6,4]
 Output:
 Original List: [2, 1, 6, 4]
 Index to be removed is: 1
 List after removing index 1 : [2, 6, 4]
 Total Number of Special elements: 1
 Explaination:
 If we delete L[1] from list : [2,6,4]
 (2+4)  =  (6)
 Here only 1 special element. So Count is 1.
 After removal of special element, list will be : [2,6,4]
 ```

````python
l = eval(input("Enter the list - "))
print("The original list is -", l)
count = 0
for i in range(len(l)):
    c = l.copy()
    c.pop(i)
    sum1 = sum(c[::2])
    sum2 = sum(c[1::2])
    if sum1 == sum2:
        print("index to be removed -",i)
        count+=1
        print("list after removing index",i,"is", l[:i] + l[i+1:])
print("Total number of special elements are - ", count)
````

Write Python Program to create a dictionary with the key as the first character and value as a list of words starting with

## WAP to find the sum of positive numbers and negative numbers in a given list using lambda function

```
input = [-4,-2,0,1,2,3,4]
output = Sum of positive numbers - 10
         Sum of negative numbers - -6
         
```


````python
positivesum = sum(list(filter(lambda x: x > 0,[-4,-2,0,1,2,3,4])))
negativesum = sum(list(filter(lambda x: x < 0,[-4,-2,0,1,2,3,4])))
print(positivesum,negativesum)
````

## PART - D
```
 Write a Python program satisfying following conditions - 
 Create a python func on countchar() that count the character of a string in a given string without using inbuilt func ons.  
For the given input, the program must print the output as shown below 
Given input string – ‘hello’
 countchar(‘l’)
 Output : 2 
 ```

````python
def countchar(s,char):
    count = 0
    for c in s:
        if c == char:
            count += 1
    return count
````

````python
countchar('hello','l')
````

## PART E

````python
def findchar(s,char):
    for i in range(len(s)):
        if s[i] == char:
            return i
    return -1
````

````python
findchar('hello', 'l')
````
