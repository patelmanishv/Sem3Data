## Inheritance 
- A class can inherit from other class
- It improves code reuse
- Constructor,attributes and methods are inherited to child class
- Parent class has no access to the child class
- child class methods overrides the parent class methods 
    - this is called method overriding
- `super()` method is used to invoke parent class methods and constructor

## Single inheritance 

````python
class Person:
    def __init__(self,name):
        self.name = name
    def show(self):
        print(self.name + " is a python teacher")
class Professor(Person):
    def display(self):
        print(f"This is {self.name}")
````

## `super() method`
- If we write constructor in both parent and child class, the parent class constructor is not available in child class
- `super()` method is used to call parent class from the child class

````python
class Phone:
    def __init__(self,price,brand,camera):
        print("Inside phone class")
        self.price = price
        self.brand = brand
        self.camera = camera
    def buy(self):
        print("Buying a phone")
class Smartphone(Phone):
    def buy(self):
        print("Buying a smartphone")
        super().buy()
        
````

````python
s = Smartphone(20000,"Iphone",13)
````

````python
s.buy()
````

## ABSTRACT CLASS
- A class derived from ABC class which belongs to abc module is known as `abstract class` in python
- ABC class is known as meta class which means class that defines behaviour of other classes.
- Abstract class can have concrete and abstract methods
- Abstract class needs to be extended and its methods implemented
- One cannot create objects of an abstract class

````python
from abc import ABC, abstractmethod

class DefenceForce(ABC):

    def __init__(self):
        self.id = 101
    
    @abstractmethod
    def area(self):
        pass
    def gun(self):
        print("The Gun is AK 47")
        
    
````

````python
class Army(DefenceForce):
    def area(self):
        print("Area is land", self.id)
````

````python
a = Army()
````

````python
a.area()
````

````python
a.gun()
````

````python
a.id
````

````python
class Navy(DefenceForce):
    def area(self):
        print("Area is Sea", self.id)
````

````python
n = Navy()
````

````python
n.area()
````

````python
n.gun()
````

## METHOD RESOLUTION ORDER IN PYTHON
- It defines an order in which the class is searched in case of multiple inheritance.
- It is based on C3 Linearization Algorithm (>= Python 2.3)


## C3 Linearization Algorithm
**Notation**
- For a list of classes - C1C2C3..Cn
  - the C1 is the head of class and the rest is tail
- C3 Linearization Algorithm is defined as MRO of sum of any class C and the linear of parents and the list of parents
```
L(C(B1,B2...BN)) = C + merge(L(B1), L(B2) ...BN))
```
- Here B1 to BN are the parent classes of C
1. Take the head of first list L(B1) from left to right
2. If the head of selected class is not in the tail of the list, add it to the linear and remove it from the list
3. Repeat the process until the classes are removed or it is impossible to find good head

## EXAMPLE DEMONSTRATION

````python
class F:
    pass
class E:
    pass
class D:
    pass
class C(D,F):
    pass
class B(D,E):
    pass
class A(B,C):
    pass
````

````python
A.mro()
````

## First step
- MRO OF D,E,F,O
- L(D) = DO
- L(E) = EO
- L(F) = FO
- L(O) = O


## Second step
```
L(C) = C + MERGE(L(D),L(F),DF)
= C + MERGE(DO,FO,DF)
= C + D + MERGE(O,FO,F)
= C + D + F + O
= CDFO
```

##  Third Step
```
L(B) = B + MERGE(L(D),L(E),DE)
= B + MERGE(DO,EO,DE)
= B + D + MERGE(O,EO,E)
= B + D + E + O
= BDEO
```

## Fourth step 
```
L(A) = A + MERGE(L(B),L(C),BC)
= A + MERGE(BDEO, CDFO,BC)
= A + B + MERGE(DEO,CDFO,C)
= A + B + C + MERGE(DEO,DFO)
= A + B + C + D + MERGE(EO,FO)
= A + B + C + D + E + F + O
= ABCDEFO
```

## OPERATOR OVERLOADING
```
__add__ = +
__sub__ = -
__mul__ = *
__truediv__ = /
```

````python
class Fraction:
    
    def __init__(self,x,y):
        self.num = x
        self.den = y
    
    def __add__(self,other):
        new_num = self.num*other.den + other.num*self.den
        new_den = self.den * other.den
        return f"{new_num}/{new_den}"
    
    def __sub__(self,other):
        new_num = self.num*other.den - other.num*self.den
        new_den = self.den * other.den
        return f"{new_num}/{new_den}"
    
    def __mul__(self,other):
        new_num = self.num * other.num
        new_den = self.den * other.den
        return f"{new_num}/{new_den}"
    
    def __truediv__(self,other):
        new_num = self.num * other.den
        new_den = self.den * other.num
        return f"{new_num}/{new_den}"
    
    def __str__(self):
        return f"{self.num} / {self.den}"
    
    def convert_decimal(self):
        return self.num / self.den
    
        
````

````python
Fraction1 = 1/3
Fraction2 = 2/5
````

````python
f1 = Fraction(1,3)
f2 = Fraction(2,5)
````

````python
f1 + f2
````

````python
f1*f2
````

````python
f1 / f2
````

````python
print(f1)
print(f1.convert_decimal())
````
